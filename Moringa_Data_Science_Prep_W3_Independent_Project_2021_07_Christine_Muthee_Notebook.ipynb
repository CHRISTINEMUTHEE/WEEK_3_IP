```python
#Which ones were the most used cities for the three days?
#most used city in the first day.
#merging df 1 with cells geo
dataset1_df=pd.concat([my_final_dataset_1,my_final_geo_df], axis=1)
dataset1_df
merged_day1=my_final_dataset_1.join(my_final_geo_df,how='outer')
merged_day1
#cleanining null values remaining
our_null=float('NaN')
merged_day1.replace("",our_null,inplace=True)
merged_day1.dropna()
# The most used city in the first day
cities_day1=merged_day1.groupby('VILLES').count()
cities_day1
#Sorting by value
cities_day1.sort_values(by=['VALUE'],ascending=0)
```

```python
#most used city in the third day.
#merging df 2 with cells geo
dataset2_df=pd.concat([my_final_dataset_2,my_final_geo_df], axis=1)
dataset2_df
merged_day2=my_final_dataset_2.join(my_final_geo_df,how='outer')
merged_day2
#cleanining null values remaining
our_null=float('NaN')
merged_day2.replace("",our_null,inplace=True)
merged_day2.dropna()
# The most used city in the third day
cities_day2=merged_day2.groupby('VILLES').count()
cities_day2
#Sorting by value
cities_day2.sort_values(by=['VALUE'],ascending=0)
```

```python
#most used city in the third day.
#merging df 3 with cells geo
dataset3_df=pd.concat([my_final_dataset_3,my_final_geo_df], axis=1)
dataset3_df
merged_day3=my_final_dataset_3.join(my_final_geo_df,how='outer')
merged_day3
#cleanining null values remaining
our_null=float('NaN')
merged_day3.replace("",our_null,inplace=True)
merged_day3.dropna()
# The most used city in the third day
cities_day3=merged_day3.groupby('VILLES').count()
cities_day3
#Sorting by value
cities_day3.sort_values(by=['VALUE'],ascending=0)
```
